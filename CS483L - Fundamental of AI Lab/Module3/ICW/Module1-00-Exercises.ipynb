### Step 11. What is the most frequent occupation?

```python
df['occupation']
```

```python
stringer=lambda x: str(x)
lst_occupation = df['occupation'].apply(stringer).tolist()
```

```python
freq = {}
for item in lst_occupation:
    if (item in freq):
        freq[item] += 1
    else:
        freq[item] = 1
 
print(freq)    
```

```python
Keymax = max(freq, key= lambda x: freq[x])
print(Keymax)
```
