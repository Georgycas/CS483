## 2. Scatter Plot

A scatter plot is used to visualize the relationship between two numerical variables.



```python
plt.hist(np.random.rand(1000))
```

```python
#rand is sampling from uniform distribution
#randn is sampling from standard normal distribution
```

```python
# Sample data
x1 = np.random.rand(50)
x2 = np.random.rand(50)
y2 = np.random.randint(0,2,50)

# Scatter plot

plt.scatter(x1,x2,  alpha=0.6,c=y2, cmap = 'bwr',label='Data Points')
plt.title('Scatter Plot')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend();
```

## 3. Line Plot
A line plot is often used to display trends over time.

```python
print(np.sin(1))
print(np.sin(1.5))
print(np.sin(2))
```

## 4. Box Plot
A box plot (or box-and-whisker plot) is used to display the distribution of data and identify outliers.

```python
# Sample data
data1 = np.random.normal(0,1,1000)
data2 = np.random.normal(20,10,1000)
data3 = np.random.normal(10,4,1000)

# Box plot

plt.boxplot([data1,data2, data3]);
```

## 5. Pie Chart
A pie chart represents data as proportions of a whole.

```python
# Sample data
la = ['A', 'B', 'C', 'D']
sizes = [20, 30, 25, 25]

# Pie chart
plt.pie(sizes, labels=la, autopct = '%1.1f%%')
plt.title('Pie Chart');
```

## 8. Combining Multiple Plots in One Figure
You can also combine multiple plots into one figure using plt.subplot().

```python
# Set up the subplot grid
plt.figure(figsize=(10, 8))

#datasets

# Scatter plot
plt.subplot(2,2,1)
x1 = np.random.rand(50)
x2 = np.random.rand(50)
y2 = np.random.randint(0,2,50)

plt.scatter(x1,x2,  alpha=0.6,c=y2, cmap = 'bwr',label='Data Points', marker='x')
plt.title('Scatter Plot')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.legend(loc = 'upper left');

# Line plot
plt.subplot(2,2,2)

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x,y, color='r');


# Box plot

plt.subplot(2,2,3)
data1 = np.random.normal(0,1,1000)
data2 = np.random.normal(20,10,1000)
data3 = np.random.normal(10,4,1000)
plt.boxplot([data1,data2, data3]);

# Bar graph
plt.subplot(2,2,4)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
plt.bar(categories, values)
# Show all plots
plt.tight_layout()
```
